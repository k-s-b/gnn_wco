```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
```

```python
def plot_model(name,**kwargs):
    performance = []
    for r in ratio:
        value = pd.read_csv("../results/%s-result-%s-%s.csv" % (name,dataset, r))[metric].mean()
        performance.append(value)
    plt.plot(performance,**kwargs)
```

# Ndata

```python
# conig
dataset = "nData"
ratio = [float(i) for i in [1,2,5,10,20]]
metric = "Test/Pr@1"

# plots
plt.figure(figsize=(15,8))
plot_model("GNN",marker="o",label="GraphFC")
plot_model("XGB",marker="^",label="XGB")
plot_model("Tabnet",marker="v",label="Tabnet")
plt.xticks(range(len(ratio)), ratio)
plt.xlabel("Inspection Ratio", fontsize=13)
plt.ylabel("Pre@1", fontsize=13)
plt.legend(fontsize=15)
plt.show()
```
